```python
emp = pd.read_excel('payroll_dataset.xlsx', sheet_name = 'Employees')
dept = pd.read_excel('payroll_dataset.xlsx', sheet_name = 'Departments')
```

```python
emp.head()
```

```python
dept.head()
```

```python
merged = pd.merge(emp, dept, on='DepartmentID')
merged
```

```python
head_of_merge = merged.head()
head_of_merge
```

```python
dup = merged[merged.duplicated()]
dup
```

```python
Q1 = merged['Salary'].quantile(0.25)
Q3 = merged['Salary'].quantile(0.75)
IQR = Q3 - Q1
lb = Q1 - 1.5 * IQR
ub = Q3 + 1.5 * IQR
```

```python
filtered = merged[(merged['Salary']>= lb) & (merged['Salary']<= ub)]
filtered
```

```python
filtered.head()
```

```python
min_sal = filtered['Salary'].min()
max_sal = filtered['Salary'].max()
```

```python
filtered['Normalized_Salary'] = (filtered['Salary'] - min_sal) / (max_sal - min_sal)
filtered.head()
```

```python
encoded = pd.get_dummies(filtered, columns=['Location'])
encoded.head()
```
