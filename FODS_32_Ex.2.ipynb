```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('inventory_dataset.csv')
data.head(10)
```

```python
data.shape
```

Handling Null Values:

Identify the columns with null values in the dataset.

```python
nullCol = data.columns[data.isnull().any()].tolist()
nullCol
```

```python
mid = data['Purchase_Price'].median()
mid
```

Impute the missing values in the 'Purchase_Price' column with the median purchase price.

```python
data['Purchase_Price'] = data['Purchase_Price'].fillna(mid)
data.head(10)
```

```python
avg = data['Sale_Price'].mean()
avg
```

Impute the missing values in the 'Sale_Price' column with the mean sale price.

```python
data['Sale_Price'] = data['Sale_Price'].fillna(avg)
data.head(10)
```

Drop any rows that still contain null values after imputation.

```python
dropped = data.dropna(axis = 0)
dropped.head()
```

```python
dropped.shape
```

Handling Outliers

Identify outliers in the 'Sale_Price' column using the Interquartile Range (IQR) method.

```python
Q1 = data['Sale_Price'].quantile(0.25)
Q1
```

```python
Q3 = data['Sale_Price'].quantile(0.75)
Q3
```

```python
IQR = Q3-Q1
lb = Q1 - 1.5 * IQR
ub = Q3 + 1.5 * IQR
```

Remove the outliers from the dataset.

```python
outliers = ((data[['Sale_Price']]<lb) | (data[['Sale_Price']] >ub)).any(axis=1)
outliers
```

```python
data[outliers]
```

```python
cleaned = data[~outliers]
cleaned
```

Normalize the 'Quantity_In_Stock' column using Min-Max scaling after removing
outliers.

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
```

```python
data['Quantity_In_Stock_Normalized'] = scaler.fit_transform(data[['Quantity_In_Stock']])
data
```

```python
cleaned['Quantity_In_Stock_Normalized'] = scaler.fit_transform(cleaned[['Quantity_In_Stock']])
cleaned
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy='median')
cleaned['Quantity_In_Stock'] = imputer.fit_transform(cleaned[['Quantity_In_Stock']])
```

Data Cleaning

Convert the 'Quantity_In_Stock' column to integer type after imputation.

```python
cleaned['Quantity_In_Stock'] = cleaned['Quantity_In_Stock'].astype(int)
```

```python
cleaned.head()
```

Encode the 'Discontinued' column using binary encoding.

```python
data_encoded = pd.get_dummies(cleaned, columns=['Discontinued'], drop_first=False)
```

```python
data_encoded.head()
```

Save the cleaned dataset to a new CSV file.

```python
data_encoded.to_csv('cleaned_dataset.csv', index=False)
```
